## 3. ***Data Wrangling***

###Step 2. Data Cleaning

#### Play Store Data

````python

# 1. Remove duplicate rows
df1 = df1.drop_duplicates()
print("After removing duplicates:", df1.shape)
print("Total duplicate rows in Play Store data:", playstore_duplicates)
````

````python
#2. Handling Missing Values
df1.isnull().sum()
````

````python
df1['Category'].replace('1.9', 'Lifestyle', inplace=True)
````

````python
df1['Rating'].unique()
````

````python
# ratings = sorted(df1['Reviews'].dropna().unique())
# ratings = [int(x) for x in ratings]
# print(ratings)
````

````python
df1['Rating'].isnull().sum()
df1['Rating'] = df1['Rating'].replace(19, 1.9)

````

````python
df1['Rating'] = df1['Rating'].replace(1., 1.0)
df1['Rating'] = df1['Rating'].replace(2., 2.0)
df1['Rating'] = df1['Rating'].replace(3., 3.0)
df1['Rating'] = df1['Rating'].replace(4., 4.0)
df1['Rating'] = df1['Rating'].replace(5., 5.0)
````

````python
#Play Store Data : Rating
df1['Rating'] = df1['Rating'].fillna(df1['Rating'].median())
df1['Rating'].isnull().sum()
````

````python
df1['Installs'].replace('Free', '0', inplace=True)
df1['Installs'].replace('0+', '0', inplace=True)
````

````python
df1['Type'].unique()
````

````python
df1['Type'] = df1['Type'].fillna('Free')
df1['Type'] = df1['Type'].replace('0','Free')
````

````python
df1['Price']=df1['Price'].replace('Everyone','0')
````

````python
df1['Content Rating'].fillna(df1['Content Rating'].mode()[0], inplace=True)
````

````python
df1['Content Rating'].isnull().sum()
````

````python
df1['Genres'].mode()
````

````python
df1['Genres'].replace('February 11, 2018', 'Tools', inplace=True)
````

````python
df1['Current Ver'].unique()
````

````python
df1['Current Ver'] = df1['Current Ver'].fillna('1.0.0')
````

````python
df1['Android Ver'].unique()
````

````python
df1['Android Ver'] = df1['Android Ver'].replace('4.4W and up','4.4 and up')
````

````python
df1['Android Ver'] = df1['Android Ver'].fillna('4.1 and up')
````

````python
# 3. Fixing Data Types :
# Reviews : convert object into int
def convert_reviews(value):
    try:
        value = str(value)
        if 'M' in value:
            return int(float(value.replace('M', '')) * 1_000_000)
        elif 'k' in value or 'K' in value:
            return int(float(value.replace('k', '').replace('K', '')) * 1_000)
        else:
            return int(value)
    except:
        return None
df1['Reviews'] = df1['Reviews'].apply(convert_reviews)
df1['Reviews'].dtype


````

````python
# Size : convert object into int
def convert_to_numeric(value):
    try:
        value = str(value).strip()

        if value.lower() == 'varies with device':
            return np.nan

        if 'M' in value:
            return float(value.replace('M', '')) * 1_000_000

        if 'k' in value.lower():
            return float(value.replace('k', '').replace('K', '')) * 1_000

        return float(value)

    except:
        return np.nan

df1['Size'] = df1['Size'].apply(convert_reviews)
df1['Size'].dtype
df1['Size'] = df1['Size'].fillna(df1['Size'].median())
````

````python
df1['Installs'] = (
    df1['Installs']
    .str.replace('[+,]', '', regex=True)
    .apply(convert_to_numeric)
    .astype('Int64')
)
df1['Installs'].dtype

````

````python
def clean_price(value):
    try:
        value = str(value).strip()

        if value == '0' or value.lower() == 'free':
            return 0.0

        value = value.replace('$', '')

        return float(value)

    except:
        return np.nan

df1['Price'] = df1['Price'].apply(clean_price)
df1['Price'].dtype
````

````python
df1['Last Updated'] = pd.to_datetime(
    df1['Last Updated'],
    errors='coerce'
)
df1['Last Updated'].dtype

````

### Correlation Analysis

The correlation matrix shows the strength and direction of relationships between
